```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
import seaborn as sns
from matplotlib.patches import Circle, Rectangle
from matplotlib.patheffects import withStroke

plt.style.use('seaborn')
background_color = '#1e1e2e'
text_color = '#ffffff'
accent_color = '#94e2d5'
grid_color = '#45475a'

colors = ['#f38ba8', '#89b4fa', '#a6e3a1', '#fab387']
alpha_fill = 0.2
line_width = 3

agents = ["Agent A", "Agent B", "Agent C", "Agent D"]
metrics = ["Call Resolution (%)", "Average Handling Time (s)", 
          "Customer Satisfaction (%)", "Calls Handled"]

performance_data = pd.DataFrame({
    "Agent": agents,
    "Call Resolution (%)": [92, 88, 85, 90],
    "Average Handling Time (s)": [240, 270, 300, 260],
    "Customer Satisfaction (%)": [95, 91, 87, 93],
    "Calls Handled": [120, 110, 115, 125]
})

fig = plt.figure(figsize=(20, 12), facecolor=background_color)
gs = fig.add_gridspec(2, 6, width_ratios=[1, 1, 1, 1, 1, 1])

ax_radar = fig.add_subplot(gs[:, :4], polar=True)
ax_radar.set_facecolor(background_color)

normalized_data = performance_data.copy()
for col in metrics:
    normalized_data[col] = (performance_data[col] - performance_data[col].min()) / \
                          (performance_data[col].max() - performance_data[col].min()) * 100

num_metrics = len(metrics)
angles = [n / float(num_metrics) * 2 * pi for n in range(num_metrics)]
angles += angles[:1]


for i in range(20, 101, 20):
    circle = Circle((0, 0), i, transform=ax_radar.transData._b, 
                   fill=False, color=grid_color, alpha=0.3, linestyle='--', linewidth=0.5)
    ax_radar.add_artist(circle)


for idx, (agent, color) in enumerate(zip(agents, colors)):
    values = normalized_data.iloc[idx, 1:].tolist()
    values += values[:1]
    

    ax_radar.plot(angles, values, 'o-', linewidth=line_width, 
                 label=agent, color=color,
                 path_effects=[withStroke(linewidth=line_width+2, foreground=color, alpha=0.3)])
    

    ax_radar.fill(angles, values, color=color, alpha=alpha_fill)
    

    ax_radar.plot(angles, values, 'o', color=color, markersize=8,
                 path_effects=[withStroke(linewidth=3, foreground=color, alpha=0.3)])


ax_radar.set_theta_offset(pi / 2)
ax_radar.set_theta_direction(-1)
ax_radar.set_rlabel_position(0)


ax_radar.set_xticks(angles[:-1])
for label, angle in zip(metrics, angles):
    if angle in (0, pi):
        ha = 'center'
    elif 0 < angle < pi:
        ha = 'left'
    else:
        ha = 'right'
    
    ax_radar.text(angle, 110, label.replace(' (%)', '\n(%)').replace(' (s)', '\n(s)'),
                 horizontalalignment=ha, verticalalignment='center',
                 color=text_color, fontsize=12, fontweight='bold')


ax_radar.set_ylim(0, 100)
ax_radar.set_rgrids([20, 40, 60, 80], angle=0, color=grid_color, alpha=0.2,
                    fontsize=10, fontweight='bold')
ax_radar.grid(True, color=grid_color, alpha=0.2)
ax_radar.spines['polar'].set_color(grid_color)
ax_radar.spines['polar'].set_alpha(0.2)


ax_radar.set_xticklabels([])

title_text = ax_radar.set_title('Agent Performance Analysis\n', 
                               pad=30, size=24, fontweight='bold', color=text_color)
title_text.set_path_effects([withStroke(linewidth=3, foreground=accent_color, alpha=0.3)])


legend = ax_radar.legend(loc='center left', bbox_to_anchor=(1.1, 0.5),
                        frameon=True, fontsize=12)
legend.get_frame().set_facecolor(background_color)
legend.get_frame().set_alpha(0.9)
legend.get_frame().set_linewidth(2)
legend.get_frame().set_edgecolor(accent_color)
plt.setp(legend.get_texts(), color=text_color)

# Add KPI cards
kpi_data = [
    {"title": "Top Resolution", 
     "value": f"{performance_data['Call Resolution (%)'].max()}%",
     "color": colors[0]},
    {"title": "Best Handling", 
     "value": f"{performance_data['Average Handling Time (s)'].min()}s",
     "color": colors[1]},
    {"title": "Peak Satisfaction", 
     "value": f"{performance_data['Customer Satisfaction (%)'].max()}%",
     "color": colors[2]},
    {"title": "Most Calls", 
     "value": f"{performance_data['Calls Handled'].max()}",
     "color": colors[3]}
]

for idx, kpi in enumerate(kpi_data):
    ax = fig.add_subplot(gs[idx//2, 4 + idx%2])
    ax.set_facecolor(background_color)

    value_text = ax.text(0.5, 0.6, kpi["value"], 
                        ha='center', va='center',
                        fontsize=24, fontweight='bold', color=kpi["color"])
    value_text.set_path_effects([withStroke(linewidth=3, foreground=kpi["color"], alpha=0.3)])

    ax.text(0.5, 0.3, kpi["title"], 
            ha='center', va='center',
            fontsize=12, color=text_color)

    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)

insights_text = (
    "PERFORMANCE INSIGHTS\n\n"
    " Agent A leads with 95% satisfaction rate\n"
    " Agent D manages highest volume: 125 calls\n"
    " Agent B shows balanced performance\n"
    " Agent C has shortest handling time"
)

fig.text(0.02, 0.02, insights_text,
         fontsize=11, color=text_color,
         bbox=dict(facecolor=background_color,
                  edgecolor=accent_color,
                  linewidth=2,
                  alpha=0.9,
                  pad=10))

plt.tight_layout()
plt.show()
```

# Customer Satisfaction

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import matplotlib.patheffects as path_effects
from matplotlib.patches import Rectangle, Circle

# Set style and colors
plt.style.use('seaborn-v0_8-dark')
background_color = '#0f172a'  # Dark blue background
text_color = '#e2e8f0'       # Light text
primary_color = '#38bdf8'    # Bright blue
secondary_color = '#818cf8'  # Purple
accent_color = '#34d399'     # Green
grid_color = '#1e293b'       # Subtle grid

# Data generation
np.random.seed(42)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
satisfaction_scores = np.random.uniform(80, 95, size=12)
variability = np.random.uniform(1, 3, size=12)
wait_times = np.random.uniform(2, 10, size=12)
call_resolutions = np.random.uniform(70, 95, size=12)
bubble_sizes = satisfaction_scores * 10

# Create main figure with grid system
fig = plt.figure(figsize=(20, 12), facecolor=background_color)
gs = GridSpec(2, 2, figure=fig, height_ratios=[1, 1.2], width_ratios=[1.5, 1])

# 1. Trend Analysis Plot
ax_trend = fig.add_subplot(gs[0, :])
ax_trend.set_facecolor(background_color)

# Create gradient effect for line
from matplotlib.collections import LineCollection
points = np.array([range(len(months)), satisfaction_scores]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
norm = plt.Normalize(satisfaction_scores.min(), satisfaction_scores.max())
lc = LineCollection(segments, cmap='cool', norm=norm, linewidth=3)
lc.set_array(satisfaction_scores)
line = ax_trend.add_collection(lc)

# Add confidence interval with gradient
gradient = np.linspace(0, 1, 256).reshape(1, -1)
gradient = np.vstack((gradient, gradient))
ax_trend.imshow(gradient, aspect='auto', extent=[0, len(months)-1, 
                satisfaction_scores.min()-variability.max(), 
                satisfaction_scores.max()+variability.max()],
                cmap='cool', alpha=0.1)

ax_trend.fill_between(
    range(len(months)),
    satisfaction_scores - variability,
    satisfaction_scores + variability,
    color=primary_color,
    alpha=0.1,
    label="Confidence Range"
)

# Add points with glow effect
scatter = ax_trend.scatter(range(len(months)), satisfaction_scores, 
                         c=satisfaction_scores, cmap='cool', 
                         s=100, zorder=5)
ax_trend.scatter(range(len(months)), satisfaction_scores, 
                c='white', s=30, zorder=6)

# Annotations for key points
max_idx = np.argmax(satisfaction_scores)
min_idx = np.argmin(satisfaction_scores)
for idx, label in zip([max_idx, min_idx], ['Peak', 'Valley']):
    ax_trend.annotate(
        f"{label}\n{satisfaction_scores[idx]:.1f}%",
        xy=(idx, satisfaction_scores[idx]),
        xytext=(10, 20 if label == 'Peak' else -20),
        textcoords='offset points',
        color=accent_color if label == 'Peak' else secondary_color,
        fontweight='bold',
        bbox=dict(
            facecolor=background_color,
            edgecolor=accent_color if label == 'Peak' else secondary_color,
            alpha=0.8,
            boxstyle='round,pad=0.5'
        ),
        arrowprops=dict(
            arrowstyle='fancy',
            color=accent_color if label == 'Peak' else secondary_color,
            alpha=0.6,
            connectionstyle="arc3,rad=0.2"
        )
    )

# Customize trend plot
ax_trend.set_title("Customer Satisfaction Evolution", 
                   fontsize=20, fontweight="bold", 
                   color=text_color, pad=20)
ax_trend.set_xlabel("Month", fontsize=12, color=text_color)
ax_trend.set_ylabel("Satisfaction Score (%)", fontsize=12, color=text_color)
ax_trend.set_xticks(range(len(months)))
ax_trend.set_xticklabels(months)
ax_trend.grid(color=grid_color, linestyle='--', linewidth=0.5)
ax_trend.set_axisbelow(True)

# 2. Bubble Chart
ax_bubble = fig.add_subplot(gs[1, 0])
ax_bubble.set_facecolor(background_color)

# Create bubble plot with enhanced styling
scatter = ax_bubble.scatter(
    wait_times, 
    call_resolutions, 
    s=bubble_sizes * 5, 
    alpha=0.6, 
    c=satisfaction_scores,
    cmap='cool',
    edgecolor=text_color,
    linewidth=1
)

# Add labels for selected points
top_3_idx = np.argsort(satisfaction_scores)[-3:]
for idx in top_3_idx:
    ax_bubble.annotate(
        f"{months[idx]}\n{satisfaction_scores[idx]:.1f}%",
        xy=(wait_times[idx], call_resolutions[idx]),
        xytext=(5, 5),
        textcoords='offset points',
        color=text_color,
        fontsize=10,
        fontweight='bold'
    )

# Customize bubble chart
ax_bubble.set_title("Satisfaction Metrics Correlation", 
                    fontsize=16, fontweight="bold", 
                    color=text_color, pad=20)
ax_bubble.set_xlabel("Average Wait Time (minutes)", fontsize=12, color=text_color)
ax_bubble.set_ylabel("Call Resolution Rate (%)", fontsize=12, color=text_color)
ax_bubble.grid(color=grid_color, linestyle='--', linewidth=0.5)

# Custom colorbar
cbar = plt.colorbar(scatter, ax=ax_bubble)
cbar.set_label("Satisfaction Score (%)", color=text_color, fontsize=12)
cbar.ax.yaxis.set_tick_params(color=text_color)
plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color=text_color)

# 3. Key Metrics Panel
ax_metrics = fig.add_subplot(gs[1, 1])
ax_metrics.set_facecolor(background_color)

# Calculate key metrics
metrics = {
    "Average Satisfaction": f"{satisfaction_scores.mean():.1f}%",
    "Satisfaction Range": f"{satisfaction_scores.min():.1f}% - {satisfaction_scores.max():.1f}%",
    "Best Performing Month": f"{months[np.argmax(satisfaction_scores)]}",
    "Average Wait Time": f"{wait_times.mean():.1f} min",
    "Resolution Rate": f"{call_resolutions.mean():.1f}%"
}

# Create styled text boxes for metrics
for i, (label, value) in enumerate(metrics.items()):
    y_pos = 0.8 - (i * 0.15)
    # Label
    ax_metrics.text(0.1, y_pos, label,
                   color=text_color,
                   fontsize=12,
                   fontweight='bold')
    # Value
    text = ax_metrics.text(0.9, y_pos, value,
                          color=primary_color,
                          fontsize=14,
                          fontweight='bold',
                          ha='right')
    # Add glow effect
    text.set_path_effects([
        path_effects.Stroke(linewidth=3, foreground=primary_color, alpha=0.2),
        path_effects.Normal()
    ])

# Style metrics panel
ax_metrics.set_title("Performance Overview", 
                     fontsize=16, fontweight="bold", 
                     color=text_color, pad=20)
ax_metrics.axis('off')

# Add subtle border
border = Rectangle((0.05, 0.1), 0.9, 0.8,
                  fill=False,
                  color=grid_color,
                  linewidth=1,
                  linestyle='--',
                  transform=ax_metrics.transAxes)
ax_metrics.add_patch(border)

# Overall layout adjustments
plt.tight_layout()
plt.show()
```

# Predictive Analytics

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from statsmodels.tsa.arima.model import ARIMA
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import make_classification
from matplotlib.gridspec import GridSpec
import matplotlib.dates as mdates
from matplotlib.patches import Rectangle, FancyBboxPatch
from matplotlib.colors import LinearSegmentedColormap

plt.style.use('seaborn-v0_8-whitegrid')
background_color = '#ffffff'
card_color = '#f8fafc'
text_color = '#1e293b'

colors = {
    'primary': '#3b82f6',      # Bright blue
    'secondary': '#8b5cf6',    # Vibrant purple
    'accent1': '#f59e0b',      # Warm amber
    'accent2': '#10b981',      # Emerald green
    'accent3': '#ef4444',      # Coral red
    'gradient1': '#6366f1',    # Indigo
    'gradient2': '#8b5cf6',    # Purple
    'gradient3': '#ec4899',    # Pink
}

colors_gradient = ['#93c5fd', '#6366f1', '#8b5cf6']  # Blue to Purple
custom_cmap = LinearSegmentedColormap.from_list('custom', colors_gradient)

np.random.seed(42)
days = 28
historical_data = np.random.randint(200, 800, size=days)
dates = [datetime(2024, 1, 1) + timedelta(days=i) for i in range(days)]

weekly_pattern = np.sin(np.linspace(0, 4*np.pi, days)) * 50
historical_data = historical_data + weekly_pattern

model = ARIMA(historical_data, order=(2, 1, 2))
fitted_model = model.fit()
forecast_steps = 7
forecast = fitted_model.get_forecast(steps=forecast_steps)
forecast_index = [dates[-1] + timedelta(days=i + 1) for i in range(forecast_steps)]
forecast_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Create figure
fig = plt.figure(figsize=(20, 14), facecolor=background_color)
gs = GridSpec(3, 2, figure=fig, height_ratios=[1.2, 1, 0.8], 
              hspace=0.4, wspace=0.3)

ax_forecast = fig.add_subplot(gs[0, :])
ax_forecast.set_facecolor(card_color)


points = np.array([mdates.date2num(dates), historical_data]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
norm = plt.Normalize(min(historical_data), max(historical_data))
lc = LineCollection(segments, cmap=custom_cmap, norm=norm, linewidth=3)
lc.set_array(historical_data)
ax_forecast.add_collection(lc)

scatter = ax_forecast.scatter(dates, historical_data, 
                            c=historical_data,
                            cmap=custom_cmap,
                            s=100, zorder=5,
                            edgecolor='white',
                            linewidth=1)

ax_forecast.plot(forecast_index, forecast_values, 
                color=colors['accent1'], 
                linestyle='--', 
                linewidth=2.5, 
                label='Forecast')

ax_forecast.fill_between(forecast_index, 
                        confidence_intervals[:, 0],
                        confidence_intervals[:, 1],
                        color=colors['accent1'], 
                        alpha=0.15,
                        label='95% Confidence')

for idx, (date, value) in enumerate([
    (dates[-1], historical_data[-1]), 
    (forecast_index[np.argmax(forecast_values)], max(forecast_values))  
]):
    color = colors['primary'] if idx == 0 else colors['accent1']
    ax_forecast.annotate(
        f'{"Current" if idx == 0 else "Peak"}\n{int(value):,}',
        xy=(date, value),
        xytext=(10, 10),
        textcoords='offset points',
        bbox=dict(
            facecolor='white',
            edgecolor=color,
            boxstyle='round,pad=0.5',
            alpha=0.9
        ),
        color=text_color,
        fontsize=10,
        fontweight='bold'
    )

ax_forecast.set_title('Call Volume Forecast', 
                     fontsize=20, 
                     fontweight='bold', 
                     color=text_color, 
                     pad=20)
ax_forecast.set_xlabel('Date', fontsize=12, color=text_color)
ax_forecast.set_ylabel('Call Volume', fontsize=12, color=text_color)

ax_forecast.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax_forecast.xaxis.set_major_locator(mdates.DayLocator(interval=4))
plt.setp(ax_forecast.get_xticklabels(), rotation=45, ha='right')

ax_pattern = fig.add_subplot(gs[1, 0])
ax_pattern.set_facecolor(card_color)

days_per_week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
weekly_avg = np.array([np.mean(historical_data[i::7]) for i in range(7)])

bar_colors = [colors['gradient' + str(i % 3 + 1)] for i in range(7)]

bars = ax_pattern.bar(days_per_week, weekly_avg, 
                     color=bar_colors,
                     alpha=0.7,
                     width=0.6)

for bar in bars:
    height = bar.get_height()
    ax_pattern.text(bar.get_x() + bar.get_width()/2., height + 5,
                   f'{int(height):,}',
                   ha='center', va='bottom', 
                   fontsize=10,
                   color=text_color)

ax_pattern.set_title('Average Daily Call Volume', 
                    fontsize=16, 
                    fontweight='bold', 
                    color=text_color,
                    pad=20)
ax_pattern.set_ylabel('Calls', fontsize=10)
ax_pattern.grid(True, axis='y', alpha=0.2)

ax_metrics = fig.add_subplot(gs[1, 1])
ax_metrics.set_facecolor(card_color)

metrics = {
    "Average Volume": (f"{int(np.mean(historical_data)):,}", colors['primary']),
    "Peak Volume": (f"{int(np.max(historical_data)):,}", colors['secondary']),
    "Forecast Trend": (f"{((forecast_values[-1] - historical_data[-1]) / historical_data[-1] * 100):+.1f}%", colors['accent1']),
    "Weekly Pattern": (f"{int(np.std(weekly_avg)):,} calls", colors['accent2'])
}

for i, (label, (value, color)) in enumerate(metrics.items()):
    y_pos = 0.8 - (i * 0.2)

    rect = Rectangle((0.1, y_pos-0.06), 0.8, 0.15,
                    facecolor='white',
                    edgecolor=color,
                    alpha=0.8,
                    transform=ax_metrics.transAxes)
    ax_metrics.add_patch(rect)
    

    ax_metrics.text(0.15, y_pos, label,
                   color=text_color,
                   fontsize=12,
                   fontweight='bold')
    ax_metrics.text(0.85, y_pos, value,
                   color=color,
                   fontsize=14,
                   fontweight='bold',
                   ha='right')

ax_metrics.set_title('Key Metrics', 
                    fontsize=16, 
                    fontweight='bold', 
                    color=text_color)
ax_metrics.axis('off')

insights = [
    ("Volume Patterns", [
        "Highest volume on Mondays",
        "Steady mid-week trend",
        "Lower weekend volumes"
    ], colors['gradient1']),
    ("Recommendations", [
        "Optimize Monday staffing",
        "Plan for forecast peaks",
        "Monitor key trends"
    ], colors['gradient2'])
]

for i, (title, items, color) in enumerate(insights):
    ax = fig.add_subplot(gs[2, i])
    ax.set_facecolor(card_color)
    
    rect = Rectangle((0.05, 0.85), 0.9, 0.15,
                    facecolor=color,
                    alpha=0.1,
                    transform=ax.transAxes)
    ax.add_patch(rect)
    

    ax.text(0.5, 0.95, title,
            ha='center',
            fontsize=14,
            fontweight='bold',
            color=text_color)
    

    for j, item in enumerate(items):
        ax.text(0.1, 0.7 - j*0.25,
                f" {item}",
                fontsize=11,
                color=text_color)
    
    ax.axis('off')

plt.tight_layout(pad=3.0)

for ax in fig.axes:
    if hasattr(ax, 'spines'):
        for spine in ax.spines.values():
            spine.set_color('#e2e8f0')
            spine.set_linewidth(1)

plt.show()
```

# Skill Based Routing Efficiency - A Sankey Diagram

```python
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
import numpy as np
from matplotlib.patches import Rectangle, FancyBboxPatch

plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 100

plt.style.use('seaborn-v0_8-whitegrid')
fig = plt.figure(figsize=(12, 8), facecolor='#ffffff')
ax = plt.gca()
ax.set_facecolor('#ffffff')


total_calls = 1000
flows = [
    total_calls,                              
    total_calls * 0.6,                        
    -(total_calls * 0.6 * 0.8),             
    total_calls * 0.4,                        
    -(total_calls * 0.4 * 0.75),             
    -(total_calls * 0.15)                     
]

labels = [
    f'Total\n{total_calls:,}',
    f'Technical\n{int(total_calls * 0.6):,}',
    f'Resolved\n{int(total_calls * 0.6 * 0.8):,}',
    f'Billing\n{int(total_calls * 0.4):,}',
    f'Resolved\n{int(total_calls * 0.4 * 0.75):,}',
    f'Escalated\n{int(total_calls * 0.15):,}'
]

sankey = Sankey(ax=ax, unit=None, scale=0.6, offset=0.1)

flow_colors = ['#4f46e5', '#06b6d4', '#10b981', 
               '#8b5cf6', '#34d399', '#f59e0b']

sankey.add(
    flows=flows,
    labels=labels,
    orientations=[0, 1, -1, 1, -1, -1],
    pathlengths=[0.3, 0.3, 0.3, 0.3, 0.3, 0.3],
    facecolor=flow_colors[0],
    alpha=0.7
)

sankey.finish()

ax.set_title(
    " A Sankey Diagram: Customer Support Flow Analytics ",
    fontsize=16,
    fontweight="bold",
    color='#1e293b',
    pad=20
)

metrics = [
    ("Technical\n80% Resolution", (0.2, 0.2), '#06b6d4'),
    ("Billing\n75% Success", (0.8, 0.2), '#8b5cf6'),
    ("Overall\n78% Efficiency", (0.5, 0.1), '#10b981')
]

for text, pos, color in metrics:
    box = FancyBboxPatch(
        (pos[0] - 0.1, pos[1] - 0.05),
        0.2, 0.1,
        boxstyle="round,pad=0.02",
        facecolor='white',
        edgecolor=color,
        alpha=0.9,
        transform=fig.transFigure
    )
    fig.patches.append(box)
    
    plt.figtext(
        pos[0],
        pos[1],
        text,
        ha='center',
        va='center',
        fontsize=9,
        color='#1e293b',
        weight='bold'
    )

# Add performance summary
summary_box = FancyBboxPatch(
    (0.02, 0.02), 0.2, 0.12,
    boxstyle="round,pad=0.02",
    facecolor='white',
    edgecolor='#4f46e5',
    alpha=0.9,
    transform=fig.transFigure
)
fig.patches.append(summary_box)

summary_text = (
    "Key Metrics:\n"
    " 80% Tech Resolution\n"
    " 75% Billing Success\n"
    " 15% Escalation Rate"
)

plt.figtext(
    0.04,
    0.08,
    summary_text,
    fontsize=8,
    color='#1e293b'
)

# Add legend
legend_elements = [
    Rectangle((0, 0), 1, 1, facecolor='#10b981', alpha=0.7,
              label='High (>75%)'),
    Rectangle((0, 0), 1, 1, facecolor='#06b6d4', alpha=0.7,
              label='Medium (60-75%)'),
    Rectangle((0, 0), 1, 1, facecolor='#f59e0b', alpha=0.7,
              label='Low (<60%)')
]

legend = ax.legend(
    handles=legend_elements,
    loc='upper right',
    bbox_to_anchor=(1.0, 1.0),
    title='Efficiency',
    frameon=True,
    fontsize=8
)
legend.get_frame().set_facecolor('white')
legend.get_frame().set_edgecolor('#e2e8f0')

plt.figtext(0.98, 0.02,
            "Support Analytics | 2024",
            ha='right',
            fontsize=7,
            color='#64748b',
            style='italic')

plt.tight_layout(rect=[0.02, 0.02, 0.98, 0.98])
plt.show()
```
